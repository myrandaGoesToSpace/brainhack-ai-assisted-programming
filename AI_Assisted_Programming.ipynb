## Introduction

Hackathons (and programming in general) involve a *lot* of debugging - writing code as fast as possible, running into errors, and spending hours trying to fix those errors. Many times, project teams find that they spend more time chasing errors in their code than actually focusing on the scientific work of the project!

**But** this workflow is about to change - with AI. Specifically, large language models (LLMs), which have shown an incredible ability to write, read, and even debug code. In the past few months, there has been a large shift toward a new workflow: AI-Assisted Programming. This workflow allows you to spend more time planning the *purpose* of your code while allowing a LLM to go through the drudgery of debugging.


Read about [this professor's experience](https://www.vanderbilt.edu/datascience/2023/04/10/chatgpt-an-ai-assistant-that-bridges-the-gap-for-novice-coders-in-the-study-of-birdsongs/) implementing a birdsong classification model mere hours after attending an AI workshop at Vanderbilt!

### What are LLMs?

Before you use these tools, it's vital to know what they are, how and why they were made, and how they get their results.

**Generative AI**

You may have heard this term before, as these are exactly the LLMs we will be using today. The "Generative" part of "Generative AI" (and also the "G" in "GPT") means that the model is trained to *generate* text. How? By predicting the next most probably word (technically, the next *token*).

While we don't know the exact training data each model used, we do know that a lot of early LLMs were trained on text available freely on the web (Wikipedia).

<u>Keep in mind:</u>

- An LLM'ss No. 1 job is to *generate text* regardless of factual correctness

## AI Assisted Workflow

1. **Prompt Engineer** your question (specifically for coding)
2. **Give feedback** if the answer is not optimal. You can even give the model more information including details on the project, documentation of a package, and example of what you want, etc.
3. **Run the code**, and send any errors to the model

  a) rinse and repeat

4. **Use** the working code. Move on to next task and start back at step (1).

Now you can get more work done, faster, and without cursing over package documentation.

```python
# Import necessary modules
from nilearn import datasets
import numpy as np

# Fetch the ADHD Dataset
n_subjects = 5  # DEMO: CHANGED TO 5 FOR DEMO PURPOSES
adhd_dataset = datasets.fetch_adhd(n_subjects=n_subjects)

# Explore the Dataset
print("Dataset keys:", adhd_dataset.keys())
print("Number of subjects:", len(adhd_dataset.func))
print("Functional MRI data files:", adhd_dataset.func)  # fMRI data
print("Confounds files:", adhd_dataset.confounds)  # Confounds data

# Convert recarray to a standard numpy array
# We need to handle structured arrays (recarrays) properly
phenotypic_data = np.array(adhd_dataset.phenotypic.tolist())
print("Sample of phenotypic data:\n", phenotypic_data[:5])

```

```python

# Step 1: Install and Import Required Libraries
# !pip install nilearn numpy scipy
from nilearn import datasets
from nilearn import plotting
from nilearn.input_data import NiftiLabelsMasker
from nilearn.connectome import ConnectivityMeasure
import numpy as np

# Step 2: Fetch the ADHD Dataset
n_subjects = 5  # Adjust the number of subjects as needed
adhd_dataset = datasets.fetch_adhd(n_subjects=n_subjects)

# Preparing the phenotypic data for ADHD/control comparison
phenotypic = np.array(adhd_dataset.phenotypic.tolist())
adhd_labels = phenotypic[:, 22]  # FROM DOCUMENTATION: 22 COLUMN IS ADHD DIAGNOSIS

# Step 3: Preprocess the fMRI Data
# Load the atlas
atlas_dataset = datasets.fetch_atlas_aal()
atlas_filename = atlas_dataset.maps
labels = atlas_dataset.labels

# Initialize the NiftiLabelsMasker with the atlas
masker = NiftiLabelsMasker(labels_img=atlas_filename, standardize=True, high_pass=0.01, t_r=2.5)

# Initialize a list to store the time series
time_series_list = []

# Loop through each subject's data
for func_file, confounds_file in zip(adhd_dataset.func, adhd_dataset.confounds):
    # Extracting the time-series from the functional images
    time_series = masker.fit_transform(func_file, confounds=confounds_file)
    time_series_list.append(time_series)

# Step 4: Extract Connectivity Measures
# Initialize a connectivity measure object
connectivity_measure = ConnectivityMeasure(kind='correlation')

# Compute connectivity matrices for each subject
correlation_matrices = connectivity_measure.fit_transform(time_series_list)

# `correlation_matrices` now holds the connectivity data for each subject

```

```python
import matplotlib.pyplot as plt
from nilearn import plotting

# Assuming `correlation_matrices` is the variable from the previous steps
# and contains the connectivity matrices for all subjects

# Select the connectivity matrix for the first subject
first_subject_connectivity_matrix = correlation_matrices[0]

# Plotting the connectivity matrix
plt.figure(figsize=(10, 10))
plt.imshow(first_subject_connectivity_matrix, vmin=-1., vmax=1., cmap='RdBu_r')
plt.colorbar()
plt.title('Connectivity Matrix for the First Subject')
plt.xlabel('Regions')
plt.ylabel('Regions')
plt.show()

# Additionally, for a more brain-oriented visualization, you can use nilearn's plotting function
# You will need the coordinates of the regions from the atlas
atlas_coords = plotting.find_parcellation_cut_coords(labels_img=atlas_dataset.maps)

# Plot the connectome for the first subject
plotting.plot_connectome(first_subject_connectivity_matrix, atlas_coords,
                         title='Connectome for the First Subject', edge_threshold="80%", node_size=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from nilearn import plotting

# Assuming `correlation_matrices` is your array of connectivity matrices
# and `adhd_labels` contains the diagnosis (1 for ADHD, 0 for control)

adhd_matrices = correlation_matrices[adhd_labels.astype("int") == 1]
control_matrices = correlation_matrices[adhd_labels.astype("int") == 0]

# Compute the average connectivity matrix for each group
avg_adhd_matrix = np.mean(adhd_matrices, axis=0)
avg_control_matrix = np.mean(control_matrices, axis=0)

# Function to plot the connectivity matrix
def plot_connectivity_matrix(matrix, title):
    plt.figure(figsize=(10, 10))
    plt.imshow(matrix, vmin=-1., vmax=1., cmap='RdBu_r')
    plt.colorbar()
    plt.title(title)
    plt.xlabel('Regions')
    plt.ylabel('Regions')
    plt.show()

# Plot the average connectivity matrices
plot_connectivity_matrix(avg_adhd_matrix, 'Average Connectivity Matrix - ADHD')
plot_connectivity_matrix(avg_control_matrix, 'Average Connectivity Matrix - Control')

# For connectome visualization, you need the coordinates of the regions
# Assuming you have `atlas_coords` from your atlas

# Function to plot the connectome
def plot_connectome(matrix, title):
    plotting.plot_connectome(matrix, atlas_coords,
                             title=title, edge_threshold="80%", node_size=20)
    plt.show()

# Plot the connectomes for both groups
plot_connectome(avg_adhd_matrix, 'Connectome - ADHD')
plot_connectome(avg_control_matrix, 'Connectome - Control')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


# Flattening the upper triangle of the connectivity matrices
adhd_upper = np.array([np.triu(m, 1).flatten() for m in adhd_matrices])
control_upper = np.array([np.triu(m, 1).flatten() for m in control_matrices])

# Prepare the dataset
X = np.vstack([adhd_upper, control_upper])
y = np.array([1] * len(adhd_upper) + [0] * len(control_upper))

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the logistic regression classifier
clf = LogisticRegression(max_iter=1000)

# Fit the classifier
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate the classifier
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
